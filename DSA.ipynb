## The process of put pivot in right position is called partitioning.
### Two types of partitoin schemes

- Hoare partitoin :
  Hoare partion the left most or right most is pivot and also the start and end pointer
- Lomuto partition :
  End element is pivot start element is called p index (partition index)

```python
def swap(start, end, element):
    if start != end:
        element[start],element[end] = element[end], element[start]
    
def partition(element, start, end):
    pivot_index = start
    pivot = element[pivot_index]

    while start < end:
        
        while start < len(element)-1 and element[start] <= pivot:
            start += 1
            
        while element[end] > pivot:
            end -= 1
            
        if start < end:
            swap(start, end, element)

    swap(pivot_index, end , element)
    
    return end
    
    

def quicksort(element, start, end):
    if start < end:
        pi = partition(element, start, end)
        quicksort(element, start, pi -1)
        quicksort(element, pi+1, end)
        
    
    

elem = [23,23,123,4,21,4123,231]
quicksort(elem, 0, len(elem)-1)
print(elem)
```

```python
def quicksort(arr, left, right):
    if left < right:
        pi = sorting(arr, left, right)
        sorting(arr, left, pi - 1)
        sorting(arr,pi + 1,right)


def sorting(arr, left, right):
    i = left
    j = right - 1
    pivot = arr[right]

    while i < j:
        while i < right and arr[i] <  pivot:
                i += 1
        while j > left and arr[j] >=  pivot:
                j -= 1
        if i < j:
            arr[i], arr[j] = arr[j], arr[i]
            
    if arr[i] > pivot:
        arr[i], arr[right] = arr[right], arr[i]

    return i
    
```

```python
def quickSort(elements, left_index, right_index):
    if left_index < right_index:  # Base case to stop recursion
        pivotIndex = partition(elements, left_index, right_index)
        quickSort(elements, left_index, pivotIndex - 1)
        quickSort(elements, pivotIndex + 1, right_index)

def partition(elements, left_index, right_index):
    pivot = elements[right_index]
    left = left_index
    right = right_index - 1  # Initialize right to one less than the end index

    while left <= right:
        while left <= right and elements[left] < pivot:
            left += 1
            
        while left <= right and elements[right] > pivot:
            right -= 1

        if left <= right:  # Swap elements if they are in the wrong order
            elements[left], elements[right] = elements[right], elements[left]
            left += 1
            right -= 1

    # Swap the pivot element to its correct position
    elements[left], elements[right_index] = elements[right_index], elements[left]
    return left
```

```python
def quicksort(arr, left, right):
    if left < right:
        pi = partition(arr, left, right)
        quicksort(arr, left, pi - 1)
        quicksort(arr, pi + 1, right)

def partition(arr, left, right):
    pivot = arr[right]
    i = left

    for j in range(left, right ):
        if arr[j] < pivot:
            arr[i], arr[j] = arr[j], arr[i]
            i += 1
    
    arr[i], arr[right] = arr[right], arr[i]
    return i
```

```python
def quickSort(a, p, r):
    if p < r:
        q = partition(a, p, r)
        quickSort(a, p, q - 1)
        quickSort(a, q + 1, r)

def partition(a, p, r): # p is left and r is right
    x = a[r]  # Pivot element
    i = p 
    for j in range(p, r):  # Iterate from p to r-1
        if a[j] <= x:
            a[i], a[j] = a[j], a[i]
            i += 1
            
    a[i], a[r] = a[r], a[i]
    return i
```
